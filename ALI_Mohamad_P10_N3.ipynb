```python


# Charger le modèle de régression logistique et le scaler
model_reg_log = joblib.load('RESULTATS/model_reg_log.pkl')
scaler = joblib.load('RESULTATS/scaler.pkl')

# Charger les données de test
XTest = billets_test.iloc[:, 0:6]

# Standardisation des données de test XTest
XTest_std = scaler.transform(XTest)

# Faire des prédictions sur les données de test
yTest_pred = model_reg_log.predict(XTest_std)

# Ajouter les prédictions de régression logistique au DataFrame
df_reg_log_test = billets_test.copy()
df_reg_log_test['Predict'] = yTest_pred

# Afficher le DataFrame final pour vérifier les mappages
display(df_reg_log_test)

# Enregistrer le DataFrame final avec les prédictions dans un fichier CSV
df_reg_log_test.to_csv('RESULTATS/billets_avec_predictions.csv', index=False)
```
